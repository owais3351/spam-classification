# Kindle Review Sentiment Analysis

# About Dataset

Context This is a small subset of dataset of Book reviews from Amazon Kindle Store category.

Content 5-core dataset of product reviews from Amazon Kindle Store category from May 1996 - July 2014. Contains total of 982619 entries. Each reviewer has at least 5 reviews and each product has at least 5 reviews in this dataset.


## Columns

asin - ID of the product, like B000FA64PK

helpful - helpfulness rating of the review - example: 2/3.

overall - rating of the product.

reviewText - text of the review (heading).

reviewTime - time of the review (raw).

reviewerID - ID of the reviewer, like A3SPTOKDG7WBLN

reviewerName - name of the reviewer.

summary - summary of the review (description).

unixReviewTime - unix timestamp.

Acknowledgements This dataset is taken from Amazon product data, Julian McAuley, UCSD website. http://jmcauley.ucsd.edu/data/amazon/

License to the data files belong to them.

Sentiment analysis on reviews.
Understanding how people rate usefulness of a review/ What factors influence helpfulness of a review.
Fake reviews/ outliers.
Best rated product IDs, or similarity between products based on reviews alone (not the best idea ikr).

# preprocessing and cleaning

```python
# Positive reviews=1==>3
# Negative reviews=0==>1,2
df['rating'] = df['rating'].apply(lambda x: 0 if x < 3 else 1)
```

```python
df['rating'].value_counts()
```

```python
# lowercase the reviews
df['reviewText'] = df['reviewText'].str.lower()
```

# data cleaning

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# Remove special characters
df['reviewText'] = df['reviewText'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))

# Remove stop words
df['reviewText'] = df['reviewText'].apply(lambda x: ' '.join([word for word in x.split() if word not in stopwords.words('english')]))

# Remove urls
df['reviewText'] = df['reviewText'].apply(lambda x: re.sub(r'http\S+|www\S+|https\S+', '', x, flags=re.MULTILINE))

# Remove extra spaces
df['reviewText'] = df['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
# remove html tags
from bs4 import BeautifulSoup
df['reviewText'] = df['reviewText'].apply(lambda x: BeautifulSoup(x, "html.parser").get_text())
```

```python
df.head()
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize the words
def lemmatize_text(text):
    return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['reviewText'] = df['reviewText'].apply(lambda x: lemmatize_text(x)).str.lower()
```
